# Trabalhando com agrupamentos em dataframes

```python
import pandas as pd
```

```python
df = pd.read_csv('../../notebooks/DATA/Universities.csv')
```

```python
df.head()
```

```python
# Temos que lembrar de colocar o parametro numeric_only, 
# se nao valores como string podem ser processados e gerar erro
df.groupby('Year').mean(numeric_only=True)
```

```python
# Podemos ordenar os resultados baseando-nos nos indices
df.groupby('Year').sum(numeric_only=True).sort_index(ascending=False)
```

```python
# Podemos passar uma lista e definir um agrupamento mais externo e um mais interno
# Os agrupamentos mais externos sao os primeiros do array de parametro
df.groupby(['Year', 'Sector']).sum()
```

```python
df.groupby('Year').describe().transpose()
```
