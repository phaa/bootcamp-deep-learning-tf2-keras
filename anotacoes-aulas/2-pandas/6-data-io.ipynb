```python
# Se eu quiser salvar o dataframe como um arquivo .csv
# Há a opcao de salvar os índices usando index=True
# Quando o arquivo for lido novamente, aparecerá como uma coluna à parte
# A não ser que voce especifique a coluna do indice
df.to_csv('output.csv', index=True)
```

```python
# Como não especificamos a coluna de indice, os indices salvos virão em uma coluna Unnamed
df = pd.read_csv('output.csv')
df
```

```python
# Para que o indice salvo seja reaproveitado, usaremos a opcao index-col
df = pd.read_csv('output.csv', index_col='Unnamed: 0')
df
```
