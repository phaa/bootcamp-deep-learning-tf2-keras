```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3]
y = [100,200,300]
```

```python
plt.plot(x,y)
plt.show()
```

```python
housing = pd.DataFrame({'rooms':[1,1,2,2,2,3,3,3],
                       'price':[100,120,190,200,230,310,330,305]})
```

```python
housing
```

```python
plt.scatter(x=housing['rooms'], y=housing['price'])
plt.show()
```

```python
# Vamos melhorar o primeiro grafico
plt.plot(x,y, color='blue', marker='o', markersize=15, linestyle='--')

plt.xlim(1,3)
plt.ylim(100,300)

plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label')

plt.show()
```
