# Visualização de dados com Seaborn

```python
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('../../dados/heart.csv')
```

```python
sns.set_theme()
sns.histplot(df['age'],kde=False,bins=50);
```

```python
df.head()
```

```python
sns.countplot(x='sex', data=df, hue='sex');
```

```python
sns.countplot(x='cp', data=df, hue='sex');
```

```python
sns.boxplot(x='target', y='thalach', data=df, hue='sex')
```

```python
sns.scatterplot(x='chol',y='trestbps',data=df,hue='sex',size='age')
```

```python
iris_df = pd.read_csv('../../dados/iris.csv')
```

```python
sns.pairplot(iris_df,hue='species')
```
